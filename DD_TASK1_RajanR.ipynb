```python
import ast




# Convert string representations of dictionaries to actual dictionaries


df_all['attributes'] =df_all['attributes'].apply(lambda x: ast.literal_eval(x) if pd.notnull(x) else {})

# Normalize the data into a DataFrame where each dictionary key becomes a column
df_attri = pd.json_normalize(df_all['attributes'])

# For nested dictionaries like 'BusinessParking', further normalize and then join
for column in df_attri.columns:
    if df_attri[column].dtype == 'object' and df_attri[column].str.startswith('{').any():
        # Normalize the nested dictionary
        nested_df = df_attri[column].apply(lambda x: ast.literal_eval(x) if pd.notnull(x) else {}).apply(pd.Series)
        # Rename columns to avoid name clashes
        nested_df.columns = [f'{column}_{nested_col}' for nested_col in nested_df.columns]
        # Join with the main DataFrame
        df_attri = df_attri.drop(column, axis=1).join(nested_df)

# Convert boolean and categorical string values into dummy variables
df_attri = df_attri.applymap(lambda x: 1 if x == 'True' or x == True else (0 if x == 'False' or x == False else x))
# Display the DataFrame
df_attri.head()
```

<h3> 6(b).Create the PCA for model_1 dataframe  </h3>
</div

- Reduce the number of dimensions from 77  to a more reasonable number that still captures the explained variance.

```python
#Create a PCA named 'df_comedy_pca' with n_pcs equal to 85 with headings of columns as PC1,PC2..
from sklearn.decomposition import PCA
# Set number of principal components for model
n_pcs = 30
model = PCA(n_components=n_pcs)
# Fit the model - return projected data
model_1_pca= model.fit_transform(model_1)
#Create a DataFrame of the projected data
model_1_pca = pd.DataFrame(model_1_pca) 
model_1_pca.head()
```

```python
#add column headings 
names = []
for x in range(n_pcs):
    names.append('PC%d'%(x+1))

model_1_pca.columns = names
model_1_pca.head()
```

```python
#Create a DataFrame named 'explained_ratio' that shows the explained_ratio and explained_ratio_sum
explained_ratio = model.explained_variance_ratio_
explained_ratio = pd.DataFrame(explained_ratio)
explained_ratio = explained_ratio.rename(columns = {0: 'explained_ratio'}) 
explained_ratio['explained_ratio_sum']= np.cumsum(explained_ratio.explained_ratio) 
explained_ratio
```

<h3> 6(e). Explaination of results

</div

-The output shows a list of restaurants along with their categories and a similarity score to the Korean Village Restaurant.Each row in the results represents a different restaurant, with its name and associated categories (such as the type of cuisine and service it offers).The number next to each restaurant (ranging between 0 and 1) is a similarity score. This score indicates how similar each listed restaurant is to the Korean Village Restaurant, based on the PCA analysis. score close to 1 suggests a high degree of similarity. For example, Seoul Korean BBQ Restaurant has a score of 0.929319, suggesting it is very similar to Korean Village Restaurant in terms of the features analyzed.
- - **What Makes Sense?**:
- Cuisine Similarity:Restaurants with Korean cuisine are ranked high in similarity to Korean Village Restaurant (e.g., Seoul Korean BBQ Restaurant, Monami, Han Corea). This aligns with expectations since customers interested in Korean Village Restaurant would likely be interested in other Korean cuisine options.
- Category Overlap:Many of the recommended restaurants share similar categories such as "Restaurants" combined with a specific cuisine (e.g., Vietnamese, Barbeque). This suggests that the system is effectively picking up on the primary category (restaurant) and secondary categories (specific cuisines).
- Geographic or Cultural Proximity:The presence of other Asian cuisine restaurants (like Vietnamese, Chinese, Japanese) in the list might reflect a geographical or cultural proximity in culinary tastes, which is a sensible basis for recommendations.
- **What is Surprising**:
- While top-ranking restaurants have scores close to Korean Village Restaurant, the decline in similarity is gradual and includes a wide variety of restaurants. It's surprising to see restaurants that seem quite different (like Greek or Indian cuisine) still having relatively high similarity scores. This could indicate a need to adjust the weighting of different features within the PCA analysis.
